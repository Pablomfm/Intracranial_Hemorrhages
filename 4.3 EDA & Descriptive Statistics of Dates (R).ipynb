## <center>INTRACRANIAL HEMORRHAGES DATABASE - EXPLORING DATA ANALYSIS (EDA) & DESCRIPTIVE STATISTICS 3 (DATES)</center>

This Notebook analyzes the relevant dates, which have been deleted in the databases shared in the repository for anonymization purposes.

1. Load data
2. Change dates into Date types
3. Descriptive statistics

### Load libraries

```r
library(data.table)
library(ggplot2)
```

### 1. Load Data

```r
df= read.csv('Databases/ICH_database_nonredudant_pseudoanonymized.csv', sep=',')
```

### 2. Change dates into Date types

```r
dates= c(4,5,7,8,11,121)
df[,dates]= lapply(df[,dates], function(x) as.Date(x, format='%Y-%m-%d'))
```

### 3. Descriptive statistics

```r
# Descriptive statistics
descriptive_dates= function (dates){
    
    # Min and max dates
    print('')
    print(paste('Min date:', min(dates)))
    print(paste('Max date:', max(dates)))
    print('')
    
    # Frequency table
    freq_tab= data.frame(table(format(dates,'%Y')))
    names(freq_tab)= c('Year', 'Freq')
    print(freq_tab)
    
    # Plot the distribution of dates
    ggplot(freq_tab, aes(x=Year, y=Freq)) + geom_bar(stat='identity', fill='steelblue')
}
```

```r
# Admission date
print('Admission date')

options(repr.plot.width=6, repr.plot.height=3)
descriptive_dates(df[['Fecha.de.ingreso']])
```

```r
# Discharge date
print('Discharge date')

options(repr.plot.width=6, repr.plot.height=3)
descriptive_dates(df[['Fecha.de.alta']])
```

```r
# CT scan date
print('CT scan date')

options(repr.plot.width=6, repr.plot.height=3)
descriptive_dates(df[['Fecha.de.TC']])
```

```r
# Blood analysis date
print('Blood analysis date')

options(repr.plot.width=6, repr.plot.height=3)
descriptive_dates(df[['Fecha.de.anlisis.de.sangre']])
```

```r
# Death date (NOTE that 1900-01-01 means that patient is still alive)
print('Death date')

options(repr.plot.width=8, repr.plot.height=4)

alive= which(df[['Fecha.de.mortalidad']] == '1900-01-01') # exclude alive people
descriptive_dates(df[['Fecha.de.mortalidad']][-alive])
```

```r
# Birth date
print('Birth date')

options(repr.plot.width=28, repr.plot.height=8)
descriptive_dates(df[['FECHA.DE.NACIMIENTO..da..mes..ao.']])
```
